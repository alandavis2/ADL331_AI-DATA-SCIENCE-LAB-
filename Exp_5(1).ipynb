```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("car data.csv")
row,col=df.shape
print("Rows ",row)
print("Column : ",col)
```

```python
df.columns
```

```python
# Find missing values for each column
missing_val = df.isna()

# Summarize the data
missing_val_count = missing_val.sum()

print("Missing value count : ", missing_val_count)
```

```python
# Create histogram of numerical column.
hist_year = df.hist(['Year'])
# OR hist_year = df.hist( column = ['Year'])
```

```python
plt.hist(df['Year'],bins=15,edgecolor='red')
plt.xlabel("Year")
plt.ylabel('Frequency')
plt.show()
```

```python
counts = df['Fuel_Type'].value_counts()
print(counts.index)
print(counts.values)
```

```python
plt.figure(figsize=(5,6))
plt.bar(counts.index,counts.values)
plt.xlabel('Fuel Type')
plt.ylabel('Frequency')
plt.show()
```
