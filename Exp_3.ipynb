```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Normal Distribution
mean = 0
sd = 1
nos = 300
n_data=np.random.normal(mean, sd, nos)
```

```python
sns.distplot(n_data)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Normal Distribution")
plt.show()
```

```python
#Uniform Distribution
lowLimit = 0
highLimit = 20
nos = 1000
uniform_data = np.random.uniform(lowLimit, highLimit, nos)
```

```python
sns.distplot(uniform_data, kde=False)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Uniform Distribution")
plt.show()
```
